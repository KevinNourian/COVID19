# Modify the Date Column

```python
# weather["date"] = weather["date"].astype("datetime64[ns]")
weather['date'] = pd.to_datetime(weather['date'], format='%m/%d/%Y')
```

# Delete Unneeded Columns

```python
weather = weather.drop(
    [
        "code",
        "max_wind_speed",
        "most_wind_direction",
        "min_temp",
        "max_temp"
    ],
    axis=1,
)
```

```python
province = province.drop(["time", "released"], axis=1)
```

# Rename Columns

```python
weather.rename(
    columns={
        "date": "Date",
        "province": "Province",
        "avg_temp": "Avg Temp",
        "precipitation": "Precipitation",
        "avg_relative_humidity": "Avg Relative Humidity"
        
    },
    inplace=True,
)
```

```python
province.rename(
    columns={
       "date": "Date",
        "province": "Province",
        "confirmed": "Confirmed",
        "deceased": "Deceased"
    },
    inplace=True,
)
province.head()
```
