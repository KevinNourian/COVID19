# Libraries

```python
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import datetime
```

# Load Data

```python
test = pd.read_csv("C:\py\Projects\TuringCollege\COVID19\DataSets\\test.csv", parse_dates =["date"], index_col ="date")
```

```python
policy = pd.read_csv(
    "C:\py\Projects\TuringCollege\COVID19\DataSets\\policy.csv",
    parse_dates=["start_date"],
    index_col = 'start_date'
)
```

# Delete Unneeded Columns

```python
test = test.drop(["time", "released", "test", "negative", "deceased"], axis=1)
```

```python
policy = policy.drop(
    [
        "policy_id",
        "country",
        "type",
        "detail",
        "end_date"
    ],
    axis=1,
)
```

# Rename Columns

```python
test.rename(
    columns={
        "confirmed": "Confirmed"  
    },
    inplace=True,
)
test.index.names = ['Date']
```

```python
policy.rename(
    columns={
        "gov_policy": "Policy",     
    },
    inplace=True,
)
```

# Rename Column Values

```python
policy.replace('Close bars and clubs', 'Close Bars and Clubs', inplace=True)
policy.replace('local government Administrative orders', 'Local Government Administrative Orders', inplace=True)
policy.replace('Close karaoke', 'Close Karaoke', inplace=True)
policy.replace('Logistics center', 'Logistics Center', inplace=True)
policy.replace('Thanks to Challenge korea', 'Thanks to Challenge Korea', inplace=True)
policy.replace('Wearing of masks', 'Wearing of Masks', inplace=True)
```

# The First Five Rows

```python
test['Count'] = test['Confirmed'].diff().fillna(0)
test.drop('Confirmed', inplace = True, axis = 1)
test
```

```python
policy.index.names = ['Date']
policy.drop_duplicates(keep="first", inplace=True)
policy.reset_index()
policy = policy.sort_values('Date', ascending = True)
```

```python
test.reset_index(inplace=True)
policy.reset_index(inplace=True)

df_outer = pd.merge(test, policy, on='Date', how='outer')

df_outer = df_outer.dropna(how='any')
df_outer['Date'] = df_outer['Date'].astype(str)
# df_outer.drop(df_outer.index[df_outer['Policy'] == "Open Data"], inplace = True)
df_outer
df_outer.drop(df_outer.index[[0,1,2,3,8,10,14,17,21]], inplace=True)
```

# Visualizations

```python
fig, ax = plt.subplots(figsize=(20, 10))

sns.set(style="whitegrid")


ax = sns.lineplot(
    x="Date",
    y="Count",
    data=df_outer,
    marker="o",
    markeredgecolor="b",
    color="crimson",
    linewidth=2.5,
    markers=True,
)

ax.annotate('Social Distancing Campaign',
            xy=('2020-02-29', 813.0),
            xycoords='data',
            xytext=(-200, 10),
            textcoords='offset points',
            arrowprops=dict(arrowstyle='->', color='black'),
            ha='center',
            va='center',
            fontsize=15)

ax.annotate('Self-Quarantine Safety Protection App',
            xy=('2020-03-07', 483.0),
            xycoords='data',
            xytext=(200, 0),
            textcoords='offset points',
            arrowprops=dict(arrowstyle='->', color='black'),
            ha='center',
            va='center',
            fontsize=15)


ax.annotate('Mandatory 14-day Self-Quarantine',
            xy=('2020-04-01', 101.0),
            xycoords='data',
            xytext=(-200, 0),
            textcoords='offset points',
            arrowprops=dict(arrowstyle='->', color='black'),
            ha='center',
            va='center',
            fontsize=15)


ax.annotate('Extends Tightened Quarantine Measures',
            xy=('2020-05-28', 79.0),
            xycoords='data',
            xytext=(-200, 0),
            textcoords='offset points',
            arrowprops=dict(arrowstyle='->', color='black'),
            ha='center',
            va='center',
            fontsize=15)



plt.title("Confirmed Cases January to June 2020", fontsize=30, y=1.05)
sns.despine(left=True, bottom=True)
plt.show();
```
