# COVID-19 in South Korea

## What is COVID-19?

Coronavirus disease (COVID-19) is an infectious disease caused by the SARS-CoV-2 virus.

Most people infected with the virus will experience mild to moderate respiratory illness and recover without requiring special treatment. However, some will become seriously ill and require medical attention. Older people and those with underlying medical conditions like cardiovascular disease, diabetes, chronic respiratory disease, or cancer are more likely to develop serious illnesses. Anyone can get sick with COVID-19 and become seriously ill or die at any age.

The virus can spread from an infected persons mouth or nose in small liquid particles when they cough, sneeze, speak, sing or breathe. These particles range from larger respiratory droplets to smaller aerosols.

Outside of China, South Korea was one of the most affected countries with COVID-19 infections.

*World Health Organization* <br>
https://www.who.int/health-topics/coronavirus

## How Did COVID-19 Start in South Korea?

**Jan 20:** The countrys first known case was a 35-year-old Chinese woman who arrived at Incheon Airport in South Korea from Wuhan, China. 

**Jan 27:** Three more known cases who were travelers from Wuhan in their 50s were identified. 

**Jan 30:** Three additional people were confirmed to have the virus. Two were from Wuhan, and one was a man who had eaten a meal at a restaurant with Patient 3 and became Patient 6, the first known case of local transmission in South Korea.

**Jan 31:**  Patient 6s wife and son were confirmed positive. Local transmission had occurred only among family members or within close social circles.

**Feb 2:**  The first case of transmission between strangers was documented when an airline passenger tested positive after sitting near an infected person on a flight.

**Feb 18:**  Patient 31, a 61-year-old woman, became the first member of Shincheonji Church of Jesus in Daegu to test positive. She reportedly ignored her symptoms, attended the church twice, lunched with friends and initially refused to be tested.

**Feb 20:**  15 more people connected to the Shincheonji church were confirmed to have the virus. A month later, thousands of people with connections to the church would test positive for the virus. The disease also appeared in hospitals, housing for the elderly and other churches, mostly near the Shincheonji church.

**Mar 25:**  The cluster at the Shincheonji church accounted for 5,080 confirmed cases of COVID-19, more than half of South Koreas total. South Korea became the second most infected country after China.

## COVID-19 Response in South Korea

Although Lithuania's population and government have major differences with those of South Korea, there are lessons we can learn. South Korea is one of very few countries to avoid both mass lockdowns and mass deaths. Unlike many countries, South Korea has seen almost no domestic backlash to its pandemic approach. Businesses largely complied with mandatory curfews. There has been no successful anti-vaccine movement. Virtually everyone wore masks as required by government guidelines. About 90% of adults have received at least one dose of the COVID-19 vaccine. During the pandemic, it has always been possible to go shopping or eat at a restaurant. These facts make it even more important to understand how the members of the Shincheonji church helped spread the virus during the period of my analysis. 

Korea Disease Control and Prevention Agency is the organization responsible for issues related to health in South Korea. After the country's biggest cluster of infections emerged among the Shincheonji church in Daegu, the South Korean government ordered testing on more than 200,000 members of the church across the country. Thousands tested positive for the virus. The government reacted quickly with other policies. This report will highlight some of the government's efforts and its effects.

## COVID-19 and Shincheonji Church

Shincheonji (pronounced *"Shin-Chan-Ji"*)was at the center of the first wave of the COVID-19 pandemic with an estimated 300,000 members. At least 5,227 people linked to its members became infected after attending a service in the city of Daegu.

The official name of the church is Shincheonji, Church of Jesus, The Temple of Tabernacle of the Testimony. It was started by Lee Man-Hee in 1984. 

According to some estimates, about 90% of the members were unvaccinated and the majority were in close contact through communal living. During their indoor meetings, members were tightly packed, helping infections to spread quickly. Shincheonji Church members were forbidden from wearing glasses and face masks during prayer services. Many of the members were in their 60s and above.

In 2020, the outbreak of COVID-19 cases in South Korea was initially centered in the city of Daegu after a 61-year-old Shincheonji member known as Patient 31 infected other church members causing the pandemic to surge in Daegu.

Eventually, half of the COVID-19 infections in Daegu and three-quarters of the infections across South Korea from January 2020 to June 2020 were linked to the members of this church.
