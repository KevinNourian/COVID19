# COVID-19 in South Korea

## What is COVID-19?

Coronavirus disease (COVID-19) is an infectious disease caused by the SARS-CoV-2 virus.

Most people infected with the virus will experience mild to moderate respiratory illness and recover without requiring special treatment. However, some will become seriously ill and require medical attention. Older people and those with underlying medical conditions like cardiovascular disease, diabetes, chronic respiratory disease, or cancer are more likely to develop serious illness. Anyone can get sick with COVID-19 and become seriously ill or die at any age. 

The virus can spread from an infected person’s mouth or nose in small liquid particles when they cough, sneeze, speak, sing or breathe. These particles range from larger respiratory droplets to smaller aerosols. 

Outside of China, South Korea was one of the most affected countries with COVID-19 infections. 

*World Health Organization* <br>
https://www.who.int/health-topics/coronavirus

## How COVID-19 is Believed to have Started in South Korea

**Jan 20:** The country’s first known case was a 35-year-old Chinese woman who arrived at Incheon Airport in South Korea from Wuhan, China. 

**Jan 27:** Within a week, there were three more known cases, all travelers from Wuhan and all in their 50s. Two were detected and isolated at airports, so it is very unlikely they passed the disease to anyone else.

**Jan 30:** Three additional people were confirmed to have the virus. Two were from Wuhan, and one was a man who had eaten a meal at a restaurant with Patient 3. He would become Patient 6, the first known case of local transmission in South Korea.

**Jan 31:**  Patient 6’s wife and son were confirmed positive, as well. Eleven days after the first case was diagnosed, there were just 10 other known cases. Local transmission had occurred only among family members or within close social circles.

**Feb 2:**  The first case of transmission between strangers was documented when an airline passenger tested positive after sitting near an infected person on a flight.

**Feb 18:**  Patient 31, a 61-year-old woman, became the first congregant at Shincheonji Church of Jesus in Daegu to test positive. It was unclear where she contracted the virus. She's been criticized for reportedly ignoring symptoms, attending the church twice, lunching with friends and initially refusing to be tested for the virus because she hadn't recently traveled abroad.

**Feb 20:**  Within two days after Patient 31 tested positive, 15 more people connected to the Shincheonji church were confirmed to have the virus. A month later, thousands of people with connections to the church would test positive for the virus. While the virus ravaged the church, more than 100 new cases had been confirmed in surrounding Daegu, as well. The disease appeared in hospitals, housing for the elderly and other churches. Most of these additional clusters were near the Shincheonji church.

**Mar 25:**  The cluster at the Shincheonji church accounted for 5,080 confirmed cases of COVID-19, more than half of South Korea’s total. South Korea became the second most infected country after China.

## COVID-19 Response in South Korea

Although the Lithuania's population and government have major difference with those of South Korea, there are lessons we can learn. South Korea is one of very few countries to avoid both mass lockdowns and mass deaths. Unlike many countries, South Korea has seen almost no domestic backlash to its pandemic approach. Businesses largely complied with mandatory curfews. There has been no successful anti-vaccine movement. Virtually everyone wore masks as required by government guidelines. About 90% of adults have received at least one dose the COVID-19 vaccine. During the pandemic it has always been possible to go shopping or eat at a restaurant. These facts make it even more important to understand how the members of Shincheonji church helped spread the virus during the period of my analysis. 

Korea Disease Control and Prevention Agency is the organization responsible for issues related to health in South Korea. After the country's biggest cluster of infections emerged among the Shincheonji church in Daegu, the South Korean government ordered testing on more than 200,000 members of the church across the country. Thousands tested positive for the virus. The government reacted quickly with other policies. This report will highlight some of government's efforts and its effects.

The South Korean response is distinct in several respects:<br>
**Early:** An early response after the first case on January 20. <br>
**Speed:** Testing was rapid and widespread. <br>
**Transparency:** Real-time and frequent information dissemination to the public. <br>
**Public-Private sector:** Enlisting companies with needed resources to fight the disease. <br>
**National organization:** Efforts were organized at the national level.

## Why were  Shincheonji Church Members Linked to the Majority of COVID19 Infections?

```python
pivot_infection = pd.pivot_table(
    data=case, index=["Infection Reason"], values="Confirmed", aggfunc="sum"
).sort_values("Confirmed", ascending=False)

total = pivot_infection["Confirmed"].sum()

pivot_province = pd.pivot_table(
    data=case, index=["Province"], values="Confirmed", aggfunc="sum"
).sort_values("Confirmed", ascending=False)

daegu = pivot_province.loc["Daegu"]
daegu = daegu["Confirmed"]


shincheonji_church_daegu = case.loc[
    (case["Province"] == "Daegu") & (case["Infection Reason"] == "Shincheonji Church")
]
shincheonji_church_daegu = shincheonji_church_daegu.set_index("Infection Reason")
shincheonji_church_daegu = shincheonji_church_daegu.loc["Shincheonji Church"]
shincheonji_church_daegu = shincheonji_church_daegu["Confirmed"]

other_daegu = daegu - shincheonji_church_daegu


shincheonji_church = pivot_infection.loc["Shincheonji Church"]
shincheonji_church = shincheonji_church["Confirmed"]

other_overall = total - shincheonji_church
```

```python
sns.set_context("paper", font_scale=2.0)

data_shincheonji_church_SK = [shincheonji_church, other_overall]
labels_shincheonji_church_SK = ["Shincheonji Church", "Other Infections"]
colors_shincheonji_church_SK = [spec_lev1, ord_color]

data_shincheonji_church_daegu = [shincheonji_church_daegu, other_daegu]
labels_shincheonji_church_daegu = ["Shincheonji Church", "Other Infections"]
colors_shincheonji_church_daegu = [spec_lev1, ord_color]

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 15))

ax1.set_title("Shincheonji Church in South Korea")
ax1.pie(
    data_shincheonji_church_SK,
    labels=labels_shincheonji_church_SK,
    colors=colors_shincheonji_church_SK,
    startangle=90,
    autopct="%0.0f%%",
    wedgeprops={"edgecolor": "black", "linewidth": 1},
    textprops={"fontsize": 15},
)

ax2.set_title("Shincheonji Church in Daegu Province")
ax2.pie(
    data_shincheonji_church_daegu,
    labels=labels_shincheonji_church_daegu,
    colors=colors_shincheonji_church_daegu,
    startangle=90,
    autopct="%0.0f%%",
    wedgeprops={"edgecolor": "black", "linewidth": 1},
    textprops={"fontsize": 15},
)


plt.tight_layout()
plt.show();
```

### Section Summary
> * The data clearly suggests that infections linked to Shincheonji Church account for about half of all infections in South Korea between January and June 2020. These infections began at the church branch in Daegu.
