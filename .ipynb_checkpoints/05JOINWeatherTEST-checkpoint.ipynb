# Libraries

```python
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from IPython.display import display

sns.set_palette("pastel")
```

# Loading Data

```python
weather = pd.read_csv(
    "C:\py\Projects\TuringCollege\COVID19\DataSets\\weather.csv",
    parse_dates=["date"],
    index_col="date",
)

test = pd.read_csv(
    "C:\py\Projects\TuringCollege\COVID19\DataSets\\test.csv",
    parse_dates=["date"],
    index_col="date",
)
```

# Display Data

```python
pd.set_option("display.max.columns", None)
```

# Delete Unneeded Columns

```python
weather = weather.drop(
    [
        "code",
        "max_wind_speed",
        "most_wind_direction",
        "min_temp",
        "max_temp"
    ],
    axis=1,
)
```

```python
test = test.drop(["time", "released", "test", "negative", "deceased"], axis=1)
```

# Rename Columns

```python
weather.rename(
    columns={
        "province": "Province",
        "avg_temp": "Avg Temp",
        "precipitation": "Precipitation",
        "avg_relative_humidity": "Avg Relative Humidity"
        
    },
    inplace=True,
)
weather.index.names = ['Date']
```

```python
test.rename(
    columns={
        "confirmed": "Confirmed"  
    },
    inplace=True,
)
test.index.names = ['Date']
```

# Column Order

```python
weather = weather[['Province', 'Avg Temp', "Precipitation", "Avg Relative Humidity"]]
```

# Dataset

```python
weather.head()
```

```python
confirmed_by_date = test.asfreq(freq='W')
confirmed_by_date
```

# Average Temp by Province

```python
avg_temp_by_province = pd.pivot_table(
    data=weather, index="Date", columns="Province", values="Avg Temp", aggfunc="mean"
)
```

```python
provinces_to_keep = ['Daegu','Gyeongsangbuk-do','Seoul','Gyeonggi-do']
avg_temp_by_province = avg_temp_by_province[provinces_to_keep]
```

```python
avg_temp_by_province = avg_temp_by_province.asfreq(freq='W')
```

```python
avg_temp_by_province.columns.name = None              
avg_temp_by_province = avg_temp_by_province.reset_index()              
```

```python
# avg_temp_by_province = avg_temp_by_province.iloc[3:]
```

```python
avg_temp_by_province = avg_temp_by_province.set_index('Date')
avg_temp_by_province
```

# Join Avg Temp and Confirmed

```python
df_inner = pd.merge(confirmed_by_date, avg_temp_by_province, on='Date', how='inner')
df_inner
```
